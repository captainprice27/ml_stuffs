Feature Selection Part

```python
movies.columns
```

```python
movies=movies[['id', 'title', 'overview', 'genre','popularity']]
# movies=movies[['id', 'title', 'overview', 'genre']]
```

```python
movies
```

```python
movies['tags'] = movies['overview']+movies['genre']
```

```python
movies
```

```python
new_data  = movies.drop(columns=['overview', 'genre'])
```

```python
new_data
```

```python
new_data.groupby('title')['popularity'].mean().sort_values(ascending=False).head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them.
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"].index[0] #demo movie
# new_data[new_data['title']=="The Godfather"].id[0]
# new_data[new_data['id']=="238"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:10]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:10]:
        print(new_data.iloc[i[0]].title)
# combining the two results before to get the recommend function
```

```python
recommend("Iron Man")
```

```python
recommend('The Godfather')
```
